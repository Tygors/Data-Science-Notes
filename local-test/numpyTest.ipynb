```python
import numpy as np
import matplotlib.pyplot as plt
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if np.random.randint(0,2) else -1
    position += step
    walk.append(position)
print(walk)
print(np.abs(walk))
plt.plot(walk)
print((np.abs(walk)>10).argmax())
```
